# 🗄️ Export Cleaned Data to MySQL

This notebook connects Python with MySQL using SQLAlchemy and exports the cleaned retail sales dataset to a MySQL database. The imported data is then verified using SQL queries.

```python
import pandas as pd
from sqlalchemy import create_engine
```

## 1. Load Cleaned Dataset

```python
df = pd.read_csv("C:/Users/DELL/OneDrive/Desktop/Retail-Sales-Analytics/data/cleaned_sales_data.csv")

df.head()
```

## 2. Connect to MySQL Database

```python
engine = create_engine(
    "mysql+pymysql://root:YOUR_PASSWORD@localhost/sales_project"
)
```

## 3. Export Dataset to MySQL

```python
df.to_sql(
    "sales_data",
    con=engine,
    if_exists="replace",
    index=False
)

print("✅ Data imported successfully!")
```

## Verify Imported Data

Display a sample of the imported records to verify that the dataset was successfully loaded into MySQL.

```python
pd.read_sql(
    "SELECT * FROM sales_data LIMIT 10;",
    engine
)
```

```python
rows = pd.read_sql(
    "SELECT COUNT(*) AS total_rows FROM sales_data;",
    engine
)

rows
```

```python
sample = pd.read_sql(
    "SELECT * FROM sales_data LIMIT 10;",
    engine
)

sample
```

## 5. SQL Analysis

```python
query = """
SELECT
    us_region,
    ROUND(SUM(revenue),2) AS Total_Revenue,
    ROUND(SUM(profit),2) AS Total_Profit
FROM sales_data
GROUP BY us_region
ORDER BY Total_Revenue DESC;
"""

pd.read_sql(query, engine)
```
